```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings, Web
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from langchain_mistralai import ChatMistralAI

from google.colab import userdata
api_key=userdata.get('MISTRAL_AI_API_KEY')


documents = SimpleDirectoryReader("data").load_data()


Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-base-en-v1.5")

llm = ChatMistralAI(model="mistral-large-latest",api_key=api_key)
Settings.llm = llm


index = VectorStoreIndex.from_documents(
    documents,
)
```

```python
query_engine = index.as_query_engine()
response = query_engine.query("summarise the documets")
print(response)
```

```python
from langchain_community.utilities import SQLDatabase
import os

with open("Chinook_sqlite.sql", "r") as f:
    sql_script = f.read()
db = SQLDatabase.from_uri("sqlite:///Chinook.db")
print(db.dialect)
print(db.get_usable_table_names())

# Split the SQL script into individual statements
sql_statements = sql_script.split(';')

# Execute each statement individually
for statement in sql_statements:
    if statement.strip():  # Skip empty statements
        db.run(statement)
# db.run("SELECT * FROM Artist LIMIT 10;")
```
