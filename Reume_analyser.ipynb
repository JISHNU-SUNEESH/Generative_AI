```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from langchain_mistralai import ChatMistralAI
```

```python
from google.colab import userdata
api_key=userdata.get('MISTRAL_AI_API_KEY')
```

```python
documents=SimpleDirectoryReader('resumes').load_data()
```

```python
Settings.embed_model=HuggingFaceEmbedding(model_name="BAAI/bge-base-en-v1.5")
llm=ChatMistralAI(model="mistral-large-latest",api_key=api_key)
Settings.llm=llm
index=VectorStoreIndex.from_documents(documents)
```

```python
query_engine=index.as_query_engine()
```

```python
jd_0="""Job Title: Test Engineer

Department: Engineering/Quality Assurance

Location: [City, State]

Reports To: QA Manager / Engineering Manager

About [Company Name]

[Insert a brief and engaging description of your company, its mission, and values.]

Job Summary

We are seeking a skilled and passionate Test Engineer to join our team. In this role, you will be responsible for designing, developing, and executing test plans to ensure the quality and reliability of our software products. You will work closely with developers, product managers, and other stakeholders to identify and resolve software defects.

Responsibilities

Design and develop comprehensive test plans and test cases based on software requirements and specifications.
Execute manual and automated tests to identify software defects and regressions.
Report and track defects using bug tracking systems (e.g., Jira, Bugzilla).
Collaborate with developers to reproduce and resolve defects.
Participate in code reviews to identify potential quality issues early in the development cycle.
Contribute to the development and maintenance of automated testing frameworks.
Stay up-to-date with the latest testing methodologies and tools.
Qualifications

Bachelor's degree in Computer Science, Engineering, or a related field.
2+ years of experience in software testing or a related role.
Strong understanding of software testing principles and methodologies.
Experience with various testing types, including functional, performance, security, and regression testing.
Proficiency in at least one scripting language (e.g., Python, JavaScript).
Experience with test automation tools and frameworks (e.g., Selenium, Appium).
Excellent communication and problem-solving skills.
Bonus Points

Experience with performance testing tools (e.g., JMeter, LoadRunner).
Familiarity with CI/CD pipelines and tools (e.g., Jenkins, GitLab CI).
Experience with mobile testing.
ISTQB or other relevant certifications.
Benefits

[List the benefits your company offers, such as competitive salary, health insurance, paid time off, and professional development opportunities.]

To Apply

Please submit your resume and cover letter to [email protected]

We are an equal opportunity employer and value diversity at our company. We do not discriminate on the basis of race, religion, color, national origin, gender, sexual orientation, age, marital status, veteran status, or disability status."""
```

```python
skill_query = f"Based on the resume, what are the candidate's key skills relevant to this job description: {jd_0}?"
suitability_query = f"Based on the resume and job description {jd_0}, is the candidate suitable for the position? Explain your reasoning."
summary_query = "Summarize the candidate's strengths and suitability for the position."

skill_response = query_engine.query(skill_query)
suitability_response = query_engine.query(suitability_query)
summary_response = query_engine.query(summary_query)
```
